### Feature Transformation

```python
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
df["race/ethnicity"]=l1.fit_transform(df["race/ethnicity"])
```

### get dummies

```python
df=pd.get_dummies(df,columns=['lunch'])
```
